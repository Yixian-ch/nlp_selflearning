## Normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure with 2x2 subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

def plot_normal(ax, mu, sigma, title):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
    y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))
    ax.bar(x, y, width=0.5)
    ax.set_title(f' = {mu},  = {sigma}')
    ax.grid(True, alpha=0.3)
    
    # Set consistent axis limits
    ax.set_xlim(-6, 6)  # Set same x-axis range for all plots
    ax.set_ylim(0, 0.5)  # Set same y-axis range for all plots

# Plot different distributions
plot_normal(ax1, mu=0, sigma=1, title='Standard Normal')
plot_normal(ax2, mu=0, sigma=2, title='Wider Spread')
plot_normal(ax3, mu=-2, sigma=1, title='Shifted Left')
plot_normal(ax4, mu=2, sigma=0.5, title='Shifted Right, Narrow')

# Add common labels
fig.text(0.5, 0.04, 'X', ha='center')
fig.text(0.04, 0.5, 'Probability Density', va='center', rotation='vertical')
plt.suptitle('Comparison of Normal Distributions', fontsize=16)

# Adjust layout
plt.tight_layout()
plt.show()
```
